# **CREATING DUMMY DATA**
Create random data(dummy data) using random numbers that follow a normal distribution

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)

mean = [-3,0]
cov = [[1.0, 0.8],[0.8, 1.0]]

# Generate 500 random variables
rv = np.random.multivariate_normal(mean, cov, 500)
```

# **Visualizing Using Scatter Plot**

```python
# Visualizing the Scatter Plot
plt.figure(figsize=(8,6))
plt.scatter(rv[:,0], rv[:,1], alpha=0.6)
plt.title('Scatter Plot of Random Variables')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

# **Visualizing Using Histogram**

```python
# Visualizing Using Histogram
# Plot histograms for each dimension
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Getting x limit of the axes
min_val = np.min(rv)
max_val = np.max(rv)
x_range = (min_val, max_val)

axs[0].hist(rv[:, 0], bins=30, color='skyblue', edgecolor='black')
axs[0].set_title('Histogram of X1 (dimension 1)')
axs[0].set_xlabel('X values')
axs[0].set_ylabel('Frequency')
axs[0].set_xlim(x_range)

axs[1].hist(rv[:, 1], bins=30, color='salmon', edgecolor='black')
axs[1].set_title('Histogram of X2 (dimension 2)')
axs[1].set_xlabel('Y values')
axs[1].set_ylabel('Frequency')
axs[1].set_xlim(x_range)

plt.tight_layout()
plt.show()
```

# **Adding Data**
Create 500 new random numbers with a two-dimensional normal distribution whose mean is (0, -3) and covariance matrix.

```python
mean1 = [0,-3]
cov= [[1.0, 0.8],[0.8, 1.0]]
rv2 = np.random.multivariate_normal(mean1, cov,500)
```

Then, visualize the data for questions 1 and 4 as one scatter plot. Display the data for question 1 legend 0 and the data for question 4 as legends 1

```python
plt.figure(figsize=(8,6))
plt.scatter(rv[:,0], rv[:,1], alpha=0.6, label='0')
plt.scatter(rv2[:,0], rv2[:,1], alpha=0.6, label='1')
plt.title('Scatter Plot of Random Variables')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

# **Data Combining**
It is easier to handle the data later if you keep it together. Combine the ndarrays created in problems 1 and 4 to create an ndarray of (1000, 2). You can combine them using or np.concatenate().np.vstack()

```python
combined_rv= np.concatenate((rv, rv2), axis=0)
print('Combined Data Shape:', combined_rv.shape)
```

```python
combined_rv = np.vstack((rv,rv2))
print("Combined data shape:", combined_rv.shape)
```

# **Labeling**
For each of the 1,000 pieces of data, please label whether it was created in problem 1 or 4. For the data in problem 1 use 0 otherwise use 1.

Add a new column of labels containing 0 and 1, and create an ndarray of (1000, 3).

Datasets used for machine learning often have this kind of structure.

```python
# Creating Label Arrays
labels1 = np.zeros((500,1))
labels2 = np.ones((500,1))

# Add a New Label Column
labeled_rv = np.hstack((rv, labels1))
labeled_rv2 = np.hstack((rv2, labels2))

# COmbine into one dataset for shape (1000.3)
combined_rv = np.vstack((labeled_rv, labeled_rv2))

print('Shape of the labeled dataset:',combined_rv.shape)
print('First six rows of the new dataset:', combined_rv[:6])
```
