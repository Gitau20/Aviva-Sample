# Question 1
Number of rice grains on the 100th day

Please create a code to calculate how many grains of rice you will have received in total on the 100th day, and output the answer. Also, please create a line graph showing the trend of "the number of grains of rice received on that day" and "the cumulative number of grains of rice received up to that day" from the 1st to the 100th day. (The vertical axis is the number of grains of rice, and the horizontal axis is the number of days.)

```python
days = 100 # number of days
rice_today = 1 # Initial Grain

# Store Values
daily_rice = [] # Each day rice
cumulative_rice = [] # Collective total rice
total_rice = 0 # total rice present

# Using for loop to compute total rice in 100 days
for day in range(1, days+1):
  daily_rice.append(rice_today) # store daily rice
  total_rice += rice_today # update total rice
  cumulative_rice.append(total_rice) # store cumulative rice
  rice_today *= 2 # double rice each day

# Print the total rice received on the 100th day
print("Total rice received on the 100th day:", total_rice)
```

```python
# Plotting the Graph
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))

# Plot daily rice
plt.plot(range(1, days+1), daily_rice, color='red', marker='o', linestyle='-', label='Daily Rice Received')

# Plot cumulative rice
plt.plot(range(1, days+1), cumulative_rice, color='blue', marker='o', linestyle='-', label='Cumulative Rice Received')

plt.xlabel('Number of Days', fontsize=14)
plt.ylabel('Number of Grains of Rice', fontsize=14)
plt.title('Rice Grain Growth For 100 Days', fontsize=16)
plt.legend()
#plt.yscale("log") using log scale to smooth the lines and see the trend easily
plt.grid(True, linestyle=':')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def sorori_shinzaemon_rice(days=100):
    daily_rice = [2 ** (i - 1) for i in range(1, days + 1)]  # Rice received on each day
    cumulative_rice = np.cumsum(daily_rice)  # Cumulative total

    total_rice_100th_day = cumulative_rice[-1]
    print(f"Total rice grains received on the 100th day: {total_rice_100th_day}")

    # Plot the data
    plt.figure(figsize=(16, 5))
    plt.plot(range(1, days + 1), daily_rice, label='Daily Rice Grains', marker='o')
    plt.plot(range(1, days + 1), cumulative_rice, label='Cumulative Rice Grains', marker='s')
    plt.yscale("log")  # Use logarithmic scale for better visualization
    plt.xlabel("Day")
    plt.ylabel("Number of Rice Grains")
    plt.title("Sorori Shinzaemon Problem: Rice Grains Growth Using Logarithmic Scale")
    plt.legend()
    plt.grid()
    plt.show()

# Example usage:
sorori_shinzaemon_rice(days=100)
```

```python
import numpy as np
import matplotlib.pyplot as plt
def compute_sorori_shinzaemon(n_days=100):
    """
    Calculates the number of rice grains received over a given number of days.

    Parameters
     ----------
        n_days: int
          Number of days to receive rice (default: 100)

    Returns
     ----------
        list_n_grains : list
        List of rice grains received each day (length = n_days)
    list_total_grains : list
        List of cumulative rice grains received up to each day (length = n_days)
    """

    # Compute the number of grains received each day (exponential growth: 1, 2, 4, 8, ...)
    daily_rice = np.array([2**(i-1) for i in range(1, n_days+1)])

    # Compute the cumulative total of rice grains received
    cumulative_rice = np.cumsum(daily_rice)

    # Total grains received after the given days
    total_rice = cumulative_rice[-1]

    # Plot the trend
    plt.figure(figsize=(10, 5))
    plt.plot(range(1, n_days+1), daily_rice, label="Rice grains per day", marker='.')
    plt.plot(range(1, n_days+1), cumulative_rice, label="Cumulative rice grains", marker='.')

    plt.yscale("log")  # Log scale due to large numbers
    plt.xlabel("Days")
    plt.ylabel("Number of rice grains (log scale)")
    plt.title(f"Rice grains received per day and cumulative total ({n_days} days)")
    plt.legend()
    plt.grid(True, which="both", linestyle="--", linewidth=0.5)

    # Show the plot
    plt.show()

    return cumulative_rice

# Example usage for 30 and 81 days

rice_30_days = compute_sorori_shinzaemon(30)
rice_81_days = compute_sorori_shinzaemon(81)

# Output results
print(f"Total rice grains received in 30 days: {rice_30_days}")
print(f"Total rice grains received in 81 days: {rice_81_days}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def compute_sorori_shinzaemon(n_days=100):
    """
    Calculates the number of rice grains received over a given number of days.

    Parameters
     ----------
        n_days: int
          Number of days to receive rice (default: 100)

    Returns
     ----------
        int: Total number of rice grains received after n_days.
    """

    # Compute the number of grains received each day (exponential growth: 1, 2, 4, 8, ...)
    daily_rice = np.array([2**(i-1) for i in range(1, n_days+1)])

    # Compute the cumulative total of rice grains received
    cumulative_rice = np.cumsum(daily_rice)

    # Plot the trend
    plt.figure(figsize=(10, 5))
    plt.plot(range(1, n_days+1), daily_rice, label="Rice grains per day", marker='.')
    plt.plot(range(1, n_days+1), cumulative_rice, label="Cumulative rice grains", marker='.')

    plt.yscale("log")  # Log scale due to large numbers
    plt.xlabel("Days")
    plt.ylabel("Number of rice grains (log scale)")
    plt.title(f"Rice grains received per day and cumulative total ({n_days} days)")
    plt.legend()
    plt.grid(True, which="both", linestyle="--", linewidth=0.5)

    # Show the plot
    plt.show()

    return cumulative_rice[-1]

def calculate_survival_days(total_rice: int, num_people: int, rice_per_person_per_day: int = 4760) -> int:
    """
    Calculates how many days a given number of people can survive on a total amount of rice.

    Parameters
    ----------
        total_rice : int
            The total number of rice grains available.
        num_people : int
            The number of people consuming the rice.
        rice_per_person_per_day : int, optional
            The number of rice grains one person consumes per day (default: 20,000 grains).

    Returns
    ----------
        int: The number of days the group can survive.
    """
    total_consumption_per_day = num_people * rice_per_person_per_day
    return total_rice // total_consumption_per_day

# Example usage for 30 and 81 days
rice_30_days = compute_sorori_shinzaemon(30)
rice_81_days = compute_sorori_shinzaemon(81)

# Assume a class of 30 people
num_people = 30
survival_days_30 = calculate_survival_days(rice_30_days, num_people)
survival_days_81 = calculate_survival_days(rice_81_days, num_people)

# Output results
print(f"Total rice grains received in 30 days: {rice_30_days}")
print(f"Total rice grains received in 81 days: {rice_81_days}")
print(f"With {num_people} people, they can survive for {survival_days_30} days on 30 days' worth of rice.")
print(f"With {num_people} people, they can survive for {survival_days_81} days on 81 days' worth of rice.")
```
