# **Fisher's Iris**
This is data on 150 irises of the following three types collected on the same day in a pasture on the Gaspe Peninsula in Canada. Setosa, virgiclor, and virginica are what are called target variables or labels here. Four types of information are recorded for each flower: sepal_length, sepal_width, petal_length, and petal_width. These four types are called features.

# **Dataset Preparation**
The machine learning library scikit-learn provides the iris dataset, so we will use this.

# 1. Acquiring Data
Load the data and store the explanatory variables in a Pandas DataFrame variable.

```python
# load data
from sklearn.datasets import load_iris
import pandas as pd
iris = load_iris()
# Store explanatory variables
x = pd.DataFrame(iris.data, columns=iris.feature_names)
x.head()
```

```python
# Store the objective variable
y = pd.DataFrame(iris.target, columns=['Species'])
y.head()
```

# **2. Data Combining**
Create x.y dataframe variable by combining the prepared in 1. Combining makes it easier to handle.

```python
# Data Combining
df = pd.concat([x,y],axis=1)
df
```

# **6. Creating a Diagram**
It is important to visualize and check the data.
- Create a pie chart showing the number of samples per label (including percentages)
- Select features one by one and visualize the distribution of data by label using box plots
- Select features one by one and visualize the distribution of data by label using a violin plot

Box plots and violin plots display similar figures. Explain how they differ and what advantages each has.

```python
# Create a pie chart showing the number of samples per label (including percentages)
import matplotlib.pyplot as plt
import seaborn as sns

# Count samples per species
species_counts = df['Species'].value_counts()

# Create labels with percentage and count
labels = [f"{species} ({count} samples)\n{count / species_counts.sum():.1%}"
          for species, count in species_counts.items()]

# Plot the pie chart
fig, ax = plt.subplots(figsize=(6, 4))
ax.pie(species_counts, labels=labels, startangle=90)
ax.set_title('Distribution of Species')
ax.axis('equal')  # Keep pie circular
plt.tight_layout()
plt.show()
```

```python
# Select features one by one and visualize the distribution of data by label using box plots
for feature in iris.feature_names:
  plt.figure(figsize=(8, 6))
  sns.boxplot(x='Species', y=feature, data=df)
  plt.title(f'Distribution of {feature} by Species')
  plt.xlabel('Species')
  plt.ylabel(feature)
  plt.grid(True, linestyle='--', alpha=0.5)
  plt.tight_layout()
  plt.show()
```

```python
# Select features one by one and visualize the distribution of data by label using a violin plot
for feature in iris.feature_names:
  plt.figure(figsize=(8, 6))
  sns.violinplot(x='Species', y=feature, data=df)
  plt.title(f'Distribution of {feature} by Species')
  plt.xlabel('Species')
  plt.ylabel(feature)
  plt.grid(True, linestyle='--', alpha=0.5)
  plt.tight_layout()
  plt.show()
```

Box plots and violin plots display similar figures. Explain how they differ and what advantages each has.

Box Plots and violin plots differ in that box plots display the median, quartiles, minimum, maximum and potential outliers while violin plots combine box plots with kernel density estimation of the distribution.

# Advantages
# Box Plots
- Very clean and concise summary of key statistics

- Good for comparing medians and spread across multiple groups

- Robust to outliers and skewed data

- Easy to interpret

# Violin Plots
- Visualizes the shape of the data distribution (e.g. peaks, multimodalities)

- Better for seeing distribution patterns (e.g. skewed or bimodal data)

- Combines the statistical summary of a box plot with the detail of a KDE plot

# **7. Confirming the relationship between features**
Create a scatter plot of all combinations (6) of 2 features selected from 4 features, color-coding them by type.

```python
# Create a scatter plot of all combinations (6) of 2 features selected from 4 features, color-coding them by type.
import itertools

# List of features to plot
features = iris.feature_names

# Generate all combinations of 2 features from the list
feature_pairs = list(itertools.combinations(features, 2))

# Set up the plot grid (3x2 grid for 6 combinations)
fig, axes = plt.subplots(3, 2, figsize=(12, 12))

# Flatten axes array for easier iteration
axes = axes.flatten()

# Plot each combination on a different subplot
for i, (feature1, feature2) in enumerate(feature_pairs):
    sns.scatterplot(data=df, x=feature1, y=feature2, hue='Species', palette='Set1', ax=axes[i])
    axes[i].set_title(f"{feature1} vs {feature2}")
    axes[i].set_xlabel(feature1)
    axes[i].set_ylabel(feature2)

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```

```python
# Create a scatter plot matrix that shows all feature combinations at once.
sns.pairplot(df, hue='Species', palette='Set1')
plt.show()
```

```python
# Create a correlation coefficient matrix for the four features
df.corr(method='pearson', min_periods=1, numeric_only=False)
```

```python
# Create a heatmap of the correlation coefficient matrix (one)
sns.heatmap(df.corr(method='pearson', min_periods=1, numeric_only=False), annot=True, cmap='coolwarm')
plt.show()
```
