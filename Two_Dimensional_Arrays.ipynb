# **Finding the gradient**
Find the gradient at each x-coordinate, where the gradient is calculated as the change in y divided by the change in x. Each change amount is approximately calculated from the difference between adjacent elements.
"Note"
- To find the exact gradient at each point, we need to use limits (differentials)
- The gradient uses the interval of the x coordinate, so it is one less than the number of elements of x.

```python
dy = np.diff(y)
dx = np.diff(x) # find the difference
grad = dy/dx # find the gradient
grad
```

```python
def compute_gradient(function, x_start, x_end, step=0.1, title='function'):

    # X array
    x = np.arange(x_start, x_end+step, step)

    # y function
    y = function(x)

    # Calculate the gradient using the change
    dy = np.diff(y)
    dx = np.diff(x)
    grad = dy/dx
    x_mid = (x[:-1] + x[1:])/2

    # Plot function and its gradient
    plt.figure(figsize=(10, 4))

    # Plot 1: y = f(x)
    plt.subplot(1, 2, 1)
    plt.plot(x, y, label='Function Graph', color='blue')
    plt.title(f"{title}")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid(True)
    plt.legend()

    # Plot 2: gradient
    plt.subplot(1, 2, 2)
    plt.plot(x_mid, grad, label="Gradient (dy/dx)", color='green')
    plt.title(f"Gradient of {title}")
    plt.xlabel("x")
    plt.ylabel("Gradient")
    plt.grid(True)
    plt.legend()

    plt.tight_layout()
    plt.show()

    return grad
# Equation Examples
# 1. y = x^2
grad1 = compute_gradient(lambda x: x**2, -50, 50, title="y = x^2")

# 2. y = 2x^2 + 2^x
grad2 = compute_gradient(lambda x: 2*x**2 + 2**x, -50, 50, title="y = 2x^2 + 2^x")

# 3. y = sin(sqrt(x))
grad3 = compute_gradient(lambda x: np.sin(np.sqrt(x)), 0, 50, title="y = sin(sqrt(x))")

# Output gradients as ndarrays
print("Gradient y = x^2:\n", grad1)
print("\nGradient y = 2x^2 + 2^x:\n", grad2)
print("\nGradient y = sin(sqrt(x)):\n", grad3)
```

# **Find the minimum value**
Find the minimum value of y for the function discussed in problem 5. The range of x is the same as in problem 5.

This time, use the min and argmin methods of ndarray instead of gradient descent. The min method can get the minimum value, and the argmin method can get the index of the minimum value.

Also, please show the gradient before and after the minimum y is reached.

```python
def compute_min_and_gradients(func, x_start, x_end, step=0.1, name="Function"):
    # Generate x and y values
    x = np.arange(x_start, x_end + step, step)
    y = func(x)

    # Find minimum value and index
    min_y = y.min()
    min_index = y.argmin()
    min_x = x[min_index]

    # Compute gradient using finite difference
    dy = np.diff(y)
    dx = np.diff(x)
    gradient = dy / dx
    x_mid = (x[:-1] + x[1:]) / 2

    # Handle gradient before and after minimum safely
    if 0 < min_index < len(gradient):
        grad_before = gradient[min_index - 1]
        grad_after = gradient[min_index]
    else:
        grad_before = grad_after = None  # Edge case (minimum at boundary)

    # Print results
    print(f"--- {name} ---")
    print(f"Minimum y value: {min_y}")
    print(f"Occurs at x = {min_x}")
    if grad_before is not None:
        print(f"Gradient before min (x = {x_mid[min_index - 1]}): {grad_before}")
        print(f"Gradient after min (x = {x_mid[min_index]}): {grad_after}")
    else:
        print("Minimum is at boundary, gradient comparison not available.")
    print()

    return min_y, min_x, grad_before, grad_after


# Define the functions and names
functions = [
    (lambda x: 0.5*x +1, -50, 50, "y=0.5x + 1"),
    (lambda x: x**2, -50, 50, "y = x^2"),
    (lambda x: 2*x**2 + 2**x, -50, 50, "y = 2x^2 + 2^x"),
    (lambda x: np.sin(np.sqrt(x)), 0, 50, "y = sin(sqrt(x))")
]

# Run for all functions
results = []
for func, x_start, x_end, name in functions:
    result = compute_min_and_gradients(func, x_start, x_end, name=name)
    results.append((name,) + result)
```
