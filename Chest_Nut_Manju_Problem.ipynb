# Assumptions Used
# Chestnut Bun
Assuming that the bun is a sphere with a diameter of 6cm, the volume will be  $V = 4/3 \pi r^3$

# Solar System
Using the estimate of Heliopause (edge of the solar system, ~120 AU):
 $V = 4/3 \pi r^3$

```python
# Computing volume for the chestnut bun and solar system
import math
bun_radius = (0.06/2)  # in meters
bun_volume = (4/3) * math.pi * (bun_radius)**3

Heliopause = 120  # in AU
solar_system_radius = (149.6e9*Heliopause)  # in meters
solar_system_volume = (4/3) * math.pi * solar_system_radius**3

print(f"Volume of a chestnut bun: {bun_volume:.2e} m^3")
print(f"Volume of the solar system: {solar_system_volume:.2e} m^3")
```

```python
import numpy as np
import matplotlib.pyplot as plt
def doubling_time(initial_volume, target_volume, doubling_minutes):
    """
    Calculates the time required for an object to reach the target volume through exponential doubling.
    """
    n = np.log2(target_volume/initial_volume)
    total_time = n * doubling_minutes
    return int(np.ceil(total_time))

def simulate_program(initial_volume, target_volume, doubling_minutes, max_time=10000):

    """
    Simulates and plots the exponential growth over time.
    """
    times = np.arange(0, max_time, doubling_minutes)
    volumes = initial_volume * 2**(times / doubling_minutes)

    plt.figure(figsize=(10,5))
    plt.plot(times, volumes, label='Growth of Object', color='blue')
    plt.axhline(y=target_volume, color='r', linestyle='--', label='Target Volume')
    plt.xlabel("Time (minutes)")
    plt.ylabel("Volume")
    plt.title("Exponential Growth of Chestnut Manju")
    plt.legend()
    plt.yscale("log")  # Log scale for better visualization
    plt.grid()
    plt.show()

# Define parameters
initial_volume = bun_volume  # Approximate volume of a chestnut bun in cubic meters
solar_system_volume = solar_system_volume  # Approximate volume within Neptune's orbit in cubic meters
doubling_minutes = 5

time_needed = doubling_time(initial_volume, solar_system_volume, doubling_minutes)
print(f"Time needed to fill the solar system: {time_needed} minutes ({time_needed / 60:.2f} hours or {time_needed / (60*24):.2f} days)")

simulate_program(initial_volume, solar_system_volume, doubling_minutes, max_time=time_needed)
```
